```python
# Qual é a média de escolaridade entre as mulheres que moram no Paraná e estão entre 25 e 30 anos? 

df_pr = df.loc[df['uf'] == 'Paraná']
df_pr = df_pr.loc[df_pr['idade'] >= 25]
df_pr = df_pr.loc[df_pr['idade'] <= 30]
df_pr['anosesco'].mean()
```

```python
# Qual é a média de renda para as pessoas na região Sul do país, que estão na força de trabalho e possuem entre 25 e 35 anos? 
df_sul = df.loc[df['uf'].isin(['Paraná', 'Santa Catarina', 'Rio Grande do Sul'])]
df_sul = df_sul.loc[df_sul['idade'] >= 25]
df_sul = df_sul.loc[df_sul['idade'] <= 35]
df_sul = df_sul.loc[df_sul['trab'] == 'Pessoas na força de trabalho']
df_sul['renda'].mean()
```

```python
# Qual é a renda média das mulheres que residem na região Norte do Brasil, possuem graduação, têm idade entre 25 e 35 anos e são pretas ou pardas? 
df_norte = df.loc[df['uf'].isin(['Acre', 'Amazonas', 'Amapá', 'Roraima', 'Rondônia', 'Pará', 'Tocantins'])]
df_norte = df_norte.loc[df_norte['idade'] >= 25]
df_norte = df_norte.loc[df_norte['idade'] <= 35]
df_norte = df_norte.loc[df_norte['graduacao'] == 'Sim']
df_norte = df_norte.loc[df_norte['cor'].isin(['Preta', 'Parda'])]
df_norte['renda'].mean()
```
